```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
latency_dataset = pd.read_csv("../data/results/latencies.csv")
latency_dataset
```

```python
qos_dataset = pd.read_csv("../data/results/offloading_qos.csv")
qos_dataset
```

```python
load_dataset = pd.read_csv("../data/results/effective_load_mean.csv")
load_dataset
```

## Plot Server load

Average load of the system captured in 24h

```python
plt.figure(figsize=(20, 10))
ax = sns.lineplot(x="time", y="value", hue="variable", data = pd.melt(load_dataset, ["time"]))
ax.set(xlabel='Time (s)', ylabel='Server load (%)')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
load_dataset_mean = load_dataset.mean().drop(["time"])
load_dataset_std = load_dataset.std().drop(["time"])

load_combined = pd.concat([load_dataset_mean, load_dataset_std], axis=1)
load_combined.columns = ["mean", "std"]
load_combined
```

# Plot QoS of the offloading capacity

Metric showing the QoS of the offloading process.  
The metric is computed as: $\frac{can\_offloading}{want\_offloading}$ and the metric range is in $[0 \ldots 1]$

```python
plt.figure(figsize=(20, 10))
ax = sns.lineplot(x="time", y="value", hue="variable", data = pd.melt(qos_dataset, ["time"]))
ax.set(xlabel='Time (s)', ylabel='QoS')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
qos_dataset_mean = qos_dataset.mean().drop(["time"])
qos_dataset_std = qos_dataset.std().drop(["time"])

qos_combined = pd.concat([qos_dataset_mean, qos_dataset_std], axis=1)
qos_combined.columns = ["mean", "std"]
qos_combined
```
